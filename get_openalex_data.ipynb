```python
#### using pyalex ####
import pyalex
from pyalex import Works, Authors, Sources, Institutions, Concepts, Publishers, Funders
```

```python
pager = (
    Works()
    .filter(
        authorships={"institutions": {"country_code": "SI"}}
    )
    .paginate(per_page=200, n_max=None)
)

w = []
c = 0
for page in pager:
    w.extend(page)
    c += len(page)
    print(c)
```

```python
np.save("si_openalex.npy",w)
```
